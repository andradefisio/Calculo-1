```python
# Calculando a derivada usando python ( https://docs.sympy.org/latest/tutorial/calculus.html )

from sympy import *
import numpy as np
import math
```

```python
x, y, z = symbols('x y z')
init_printing(use_unicode = True)
```

```python
# A primeira derivada da função f(x) = y = 0.0254x^3 - 2.735x^2 + 79.24x - 176.25 é dada por f´(x):

derivada1 = diff(0.0254*x**3 - 2.735*x**2 + 79.24*x - 176.25, x)
print(derivada1)
```

```python
# Pelo gráfico, observa-se que os casos de óbitos diários crescem até por volta do 20º dia. 

# Devemos então observar o comportamento da função neste intervalo. Para analisar o crescimento, a derivada de primeira ordem
# dada por f´(x) = 0.0762*x**2 - 5.47*x + 79.24 deve ser igual a zero.
# Dessa forma, encontramos os valores das raízes ou onde a derivada de f´(x) é igual a zero. 

# Usar a fórmula de Baskar para encontrar as raízes. 

a = 0.0762
b = -5.47
c = 79.24
delta = (b**2) - (4*a*c)

x1 = (-b + math.sqrt(delta))/(2*a)
x2 = (-b - math.sqrt(delta))/(2*a)

print('A raiz x1 vale', x1)
print('A raiz x2 vale', x2)


```

As raízes encontradas, x1 e x2 são os pontos onde a derivada da f´(x) é igual a zero. 
Vamos verificar o crescimento/decrescimento da função. Para isto, vamos substituir valores de x (dias) menores que 20 em f´(x). 



```python
# obs.: rodar as células na sequencia para não alterar os valores de x. 
x = 19
```

```python
# quando o x for igual a 19, qual será o valor da derivada? > ou < que zero?
derivada_dia19 = 0.0762*x**2 - 5.47*x + 79.24

print('O valor da derivada quando o número de dias for igual a 19 é ', derivada_dia19, '\nLogo o valor da derivada é maior que zero, indicando que a função cresce.')
```

E para valores maiores que 20?
Vamos substituir em f´(x)


```python
x = 21 
```

```python
# quando o x for igual a 21, qual será o valor da derivada? > ou < que zero?
derivada_dia21 = 0.0762*x**2 - 5.47*x + 79.24

print('O valor da derivada quando o número de dias for igual a 21 é ', derivada_dia21, '\nLogo o valor da derivada é menor que zero, indicando que a função decresce.')
```

Par encontrar o ponto de inflexão, devemos encontrar a derivada de segunda ordem da f(x) ou y = 0.0254x^3 - 2.735x^2 + 79.24x - 176.25

```python
# derivada de segunda ordem de f(x)
# devemos redefinir x, porque pelos calculos anteriores, o x foi armazenado com um valor.

x, y, z = symbols('x y z')
init_printing(use_unicode = True)

derivada2 = diff(0.0254*x**3 - 2.735*x**2 + 79.24*x - 176.25, x, x)
print(derivada2)
```

```python
# A derivada de segunda ordem ou f´´(x) = 0.1524*x - 5.47
# Para determinar o ponto de inflexão, a f´´(x) deve ser = a zero.

# f´´(x) = 0.1524*x - 5.47 = 0
print('A derivada de segunda ordem f´´(x) é igual a ', 5.47/0.1524)
```

```python
# Substituindo 35.892388451443566 em 
# f´´(x) = 0.1524*x - 5.47 = 0
d = 0.1524*35.892388451443566 - 5.47
print('O valor da derivada quando x = 35.892388451443566 é ', d)
```

```python
# Além disso, o valor da derivada f´´´(x) deve ser != zero. 
x, y, z = symbols('x y z')
init_printing(use_unicode = True)
derivada3 = diff(0.0254*x**3 - 2.735*x**2 + 79.24*x - 176.25, x, x, x)

print('O valor da derivada f´´´(x) é ', derivada3, '\nComo este valor é diferente de zero, o ponto no qual x = 20 é ponto de inflexão.')
```
